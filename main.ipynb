```python
import matplotlib.pyplot as plt

# Data aktual
bulan = ['Nov 2023', 'Des 2023', 'Jan 2024', 'Feb 2024', 'Mar 2024', 'Apr 2024']
harga = [25000, 30000, 28000, 32000, 27000, 35000]

# Data prediksi
bulan_prediksi = ['Mei 2024', 'Jun 2024', 'Jul 2024']
harga_prediksi = [33000, 30000, 28000]

# Membuat grafik
plt.figure(figsize=(10, 6))
plt.plot(bulan, harga, marker='o', label='Harga Aktual')
plt.plot(bulan_prediksi, harga_prediksi, marker='o', linestyle='--', label='Prediksi Harga')
plt.xlabel('Bulan')
plt.ylabel('Harga Cabe (Rp/Kg)')
plt.title('Pergerakan Harga Cabe Merah')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

1c. Penjelasan gambar:

Grafik menampilkan pergerakan harga cabe merah dari bulan November 2023 sampai dengan April 2024 menggunakan garis padat berwarna biru dan titik-titik sebagai penanda.
Grafik juga menampilkan prediksi harga cabe merah untuk 3 bulan berikutnya (Mei, Juni, dan Juli 2024) menggunakan garis putus-putus berwarna hijau dan titik-titik sebagai penanda.
Sumbu x menunjukkan bulan, sedangkan sumbu y menunjukkan harga cabe dalam satuan Rupiah per kilogram (Rp/Kg).
Dari data aktual, terlihat bahwa harga cabe merah berfluktuasi, dengan harga tertinggi pada bulan April 2024 (Rp 35.000/Kg) dan harga terendah pada bulan November 2023 (Rp 25.000/Kg).
Prediksi harga untuk bulan Mei 2024 adalah Rp 33.000/Kg, untuk bulan Juni 2024 adalah Rp 30.000/Kg, dan untuk bulan Juli 2024 adalah Rp 28.000/Kg.
Grafik dilengkapi dengan label pada sumbu x dan y, judul grafik, dan legenda untuk membedakan data aktual dan data prediksi.


Overfitting adalah situasi di mana model machine learning terlalu kompleks dan tidak dapat melakukan generalisasi dengan baik pada data baru yang belum pernah dilihat sebelumnya. Ini terjadi ketika model "menghafalkan" pola-pola spesifik dari data pelatihan dan tidak dapat menangkap pola-pola umum yang berlaku untuk data di luar data pelatihan.

Ketika terjadi overfitting, model akan memiliki performa yang sangat baik pada data pelatihan tetapi performa yang buruk pada data uji atau data baru. Hal ini disebabkan oleh model yang terlalu rumit dan terlalu menyesuaikan diri dengan noise (noise-fitting) atau fluktuasi acak dalam data pelatihan.
Overfitting dapat menyebabkan masalah generalisasi, di mana model tidak dapat melakukan prediksi yang akurat pada data baru yang belum pernah dilihat sebelumnya. Untuk mengatasi overfitting, dapat dilakukan beberapa teknik seperti regularisasi, penambahan data pelatihan, penyederhanaan model, atau penggunaan teknik validasi silang.

Regresi Logistik Binary digunakan untuk memprediksi variabel dependen biner (nilai 0 atau 1) berdasarkan satu atau lebih variabel independen. Berikut adalah contoh penggunaan Regresi Logistik Binary beserta penjelasan gambar:

Misalkan kita ingin memprediksi apakah seorang pelanggan akan membeli produk atau tidak berdasarkan usia dan pendapatan mereka. Variabel dependen adalah "Membeli" (0 untuk tidak membeli dan 1 untuk membeli), sedangkan variabel independen adalah "Usia" dan "Pendapatan".


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
usia = np.array([25, 30, 35, 40, 45, 50, 55, 60])
pendapatan = np.array([30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000])
membeli = np.array([0, 0, 0, 1, 1, 1, 1, 1])  # 0 = tidak membeli, 1 = membeli

# Membuat plot
plt.figure(figsize=(8, 6))
plt.scatter(usia[membeli == 0], pendapatan[membeli == 0], marker='o', label='Tidak Membeli')
plt.scatter(usia[membeli == 1], pendapatan[membeli == 1], marker='+', label='Membeli')
plt.xlabel('Usia')
plt.ylabel('Pendapatan')
plt.title('Prediksi Pembelian Produk')
plt.legend()
plt.show()
```

2. Penjelasan gambar:

Sumbu x menunjukkan usia pelanggan, sedangkan sumbu y menunjukkan pendapatan pelanggan.
Titik-titik pada grafik merepresentasikan data pelanggan, dengan warna dan bentuk yang berbeda untuk pelanggan yang membeli (tanda '+' berwarna hijau) dan tidak membeli (lingkaran berwarna biru).
Dari grafik, terlihat bahwa pelanggan dengan usia yang lebih tua dan pendapatan yang lebih tinggi cenderung membeli produk, sedangkan pelanggan dengan usia yang lebih muda dan pendapatan yang lebih rendah cenderung tidak membeli produk.
Model Regresi Logistik Binary akan menggunakan informasi usia dan pendapatan untuk memprediksi probabilitas seorang pelanggan membeli produk atau tidak.
Dengan menetapkan ambang batas probabilitas tertentu (misalnya 0,5), model akan mengklasifikasikan pelanggan ke dalam kategori "membeli" atau "tidak membeli" berdasarkan probabilitas yang dihitung.

Dalam contoh ini, Regresi Logistik Binary membantu mengidentifikasi hubungan antara variabel independen (usia dan pendapatan) dengan variabel dependen (membeli atau tidak membeli produk) sehingga dapat digunakan untuk memprediksi perilaku pembelian pelanggan baru berdasarkan informasi usia dan pendapatan mereka.
